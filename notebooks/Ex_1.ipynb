```python
dataset = pd.read_csv('../models/41_25.csv', delimiter=',', header=None)
print(dataset)
```

```python
print(sum(dataset[0]))
```

```python
dataset_centralized = dataset.copy()

means = list(np.mean(dataset[i]) for i in dataset)
for i in dataset:
    dataset_centralized[i] = dataset[i] - means[i]

print(dataset_centralized)
```

```python
d_cent_transposed = dataset_centralized.transpose()
print(d_cent_transposed)
```

```python
pca = PCA(n_components = 1, svd_solver = 'full')
pca.fit(d_cent_transposed)
```

```python
pca.components_
```
