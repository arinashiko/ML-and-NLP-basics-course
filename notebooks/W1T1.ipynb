```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
input_ds = pd.read_csv('../data/W1T1/41_25.csv', delimiter=',', header=None) # read data setting values delimiter & headers absence

input_ds
```

```python
sum(input_ds[0])  # first sign for all objects sum
```

```python
input_ds_centralized = input_ds.copy()

means = list(np.mean(input_ds[i]) for i in input_ds)  # all signs average values
for i in input_ds:
    input_ds_centralized[i] = input_ds[i] - means[i]  # centralize each cell

input_ds_centralized
```

```python
sum(input_ds_centralized[0])  # centralized first sign sum equals to 6e-14 ~= 0
```

```python
pca = PCA(n_components=2, svd_solver='full')  # init PCA object with specific settings
pca.fit(input_ds_centralized)  # input centralized dataset to PCA object
data_pca = pca.transform(input_ds_centralized)  # compute PCAs
data_pca = pd.DataFrame(data_pca, columns=['PC1', 'PC2'])

data_pca
```

```python
pca.explained_variance_ratio_ # print all explained variance ratios
```

```python
sum(pca.explained_variance_ratio_[0:2]) # sum first 2 ratios
```

```python
plt.scatter(data_pca["PC1"], data_pca["PC2"]) # build plot
plt.show()
```
